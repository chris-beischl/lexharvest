```python
import json
import sqlite3

import pandas as pd
```

```python
conn = sqlite3.connect("../lexharvest.db")
conn.row_factory = sqlite3.Row

rows = conn.execute("SELECT * FROM vocab_entries WHERE status = 'done' LIMIT 30").fetchall()
df = pd.DataFrame([dict(r) for r in rows])
df["translations"] = df["translations"].apply(json.loads)
df
```
