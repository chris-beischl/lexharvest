```python
import re

import httpx
```

```python
base_url = "https://en.wiktionary.org/api/rest_v1/page/definition"

word = "intentar"
response = httpx.get(
    f"{base_url}/{word}",
    headers={"User-Agent": "lexharvest/0.1 (https://github.com/chris-beischl/lexharvest)"},
)
print(response)

es = response.json()["es"][0]
pos = es["partOfSpeech"]
defs = es["definitions"]

defs = defs[:3]
for d in defs:
    text = re.sub(r"<[^>]+>", "", d["definition"])
    print(f"{pos}: {text}")
```
